```python
from collections import defaultdict,Counter
from glob import glob
```

```python
def topic_modelling(txt):
    symbols = '"#,.?`()!<>:[]/{};#@%$&*_-'
    data = []
    val = []
    for i in txt:
        title = i.split()
        if 'Title:' in title:
            print(i)
            break


    stoplist = open('stopwords.txt','r')

    for line in stoplist:
        val.append(line.split())

    flatlist = [item for sub in val for item in sub]

    stopword = set(flatlist)       

    data = [[word for word in item.lower().split() if word not in stopword and word not in symbols and word.isalpha()] for item in txt]
    #print(data)

    frequency = defaultdict(int)

    for text in data:
        for token in text:
            frequency[token] += 1
    #print(frequency)
    
    return frequency
```

```python
path = 'Guttenberg novels/*.txt'
files = glob(path)

for i in files:
    f = open(i,'r')
    val = topic_modelling(f)
    word_frequency = Counter(val)
    for word, count in word_frequency.most_common(5):
        print(word,":",count)
    f.close()
    print('\n')
```
